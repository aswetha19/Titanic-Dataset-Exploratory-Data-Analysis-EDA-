```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Pairplot - selected numeric columns with hue='Survived'
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', diag_kind='kde')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python



# Set plot style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)

# 1. Survival rate by Sex
plt.figure()
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Sex")
plt.show()

# 2. Survival rate by Pclass
plt.figure()
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()

# 3. Survival rate by Embarked
plt.figure()
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title("Survival Rate by Embarkation Port")
plt.show()

# 4. Age distribution by Survival
plt.figure()
sns.histplot(data=df, x='Age', hue='Survived', kde=True, multiple='stack')
plt.title("Age Distribution by Survival")
plt.show()

# 5. Fare distribution by Survival
plt.figure()
sns.histplot(data=df, x='Fare', hue='Survived', kde=True, multiple='stack')
plt.title("Fare Distribution by Survival")
plt.show()

# 6. Scatterplot: Fare vs Age colored by Survival
plt.figure()
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')
plt.title("Fare vs Age by Survival")
plt.show()

# 7. Boxplots: Age & Fare by Survival
for col in ['Age', 'Fare']:
    plt.figure()
    sns.boxplot(x='Survived', y=col, data=df)
    plt.title(f"{col} by Survival")
    plt.show()
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)

# 1. Histograms for numeric columns
numeric_cols = ['Age', 'Fare']
for col in numeric_cols:
    plt.figure()
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Histogram of {col}")
    plt.show()

# 2. Boxplots for numeric columns by Survival
for col in numeric_cols:
    plt.figure()
    sns.boxplot(x='Survived', y=col, data=df)
    plt.title(f"Boxplot of {col} by Survival")
    plt.show()

# 3. Scatterplot: Fare vs Age (colored by Survival)
plt.figure()
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Scatterplot of Fare vs Age by Survival")
plt.show()

# 4. Scatterplot: Fare vs Pclass (colored by Survival)
plt.figure()
sns.scatterplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Scatterplot of Fare vs Pclass by Survival")
plt.show()
```
